```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('results.csv')

data
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['pages'], data['time_ns'], 'o-')
plt.xscale('log', base=2)  # Escala logarítmica base 2 en el eje X
plt.xlabel('Número de páginas')
plt.ylabel('Tiempo por acceso (ns)')
plt.title('Tiempo de acceso a memoria vs Número de páginas')
plt.grid(True)
plt.savefig('tlb_size.png')
plt.show()
```
